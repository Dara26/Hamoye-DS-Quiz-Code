```python
#import csv file
import numpy as np
import pandas as pd
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00471/Data_for_UCI_named.csv',error_bad_lines=False)
df.head(20)
```

```python
#drop stab column to leave stabf as sole dependent variable
df.drop('stab',inplace=True,axis=1)
df.head()
```

```python
#Split data into train and test sets
X = df.drop('stabf',axis=1)
y = df.stabf
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
#Standardize training and testing datasets
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train)
x_train_normalized = scaler.transform(x_train)
x_test_normalized = scaler.transform(x_test)
x_train_normalized_df = pd.DataFrame(x_train_normalized, columns = x_train.columns)
x_test_normalized_df = pd.DataFrame(x_test_normalized, columns = x_test.columns)
```

```python
#train a random forest classifier
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state=1)
rfc.fit(x_train_normalized_df,y_train)
rfc_pred=rfc.predict(x_test_normalized_df)
```

```python
#Question 14
#Calculate accuracy score using random forest classifier
from sklearn.metrics import accuracy_score
accuracy_rfc = accuracy_score(y_test, rfc_pred)
accuracy_rfc.round(4)
```

```python
#train an extreme boosting model
from xgboost import XGBClassifier
XGB=XGBClassifier(random_state=1)
XGB.fit(x_train_normalized_df,y_train)
XGB_pred=XGB.predict(x_test_normalized_df)
```

```python
#Question 15
#Calculate accuracy score using xgboost
accuracy_xgb = accuracy_score(y_test, XGB_pred)
accuracy_xgb.round(4)
```

```python
#train a light gradient boosting model
from lightgbm import LGBMClassifier
LGBM=LGBMClassifier(random_state=1)
LGBM.fit(x_train_normalized_df,y_train)
LGBM_pred=LGBM.predict(x_test_normalized_df)
```

```python
#Question 16
#Calculate accuracy score using lgbm
accuracy_lgbm = accuracy_score(y_test, LGBM_pred)
accuracy_lgbm.round(4)
```

```python
#train an extra trees classifier
from sklearn.ensemble import ExtraTreesClassifier
ETC=ExtraTreesClassifier(random_state=1)
ETC.fit(x_train_normalized_df,y_train)
ETC_pred=ETC.predict(x_test_normalized_df)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,ETC_pred, digits=4))
```

```python
n_estimators = [50, 100, 300, 500, 1000]

min_samples_split = [2, 3, 5, 7, 9]

min_samples_leaf = [1, 2, 4, 6, 8]

max_features = ['auto', 'sqrt', 'log2', None]
```

```python
hyperparameter_grid = {'n_estimators': n_estimators,

                       'min_samples_leaf': min_samples_leaf,

                       'min_samples_split': min_samples_split,

                       'max_features': max_features}
```

```python
#Run a randomized cross validation search using the hyper parameter grid
from sklearn.model_selection import RandomizedSearchCV
Random_inputs=RandomizedSearchCV(ExtraTreesClassifier(random_state=1),
                                hyperparameter_grid, random_state = 1,cv=5,verbose=1,
                               scoring = 'accuracy', n_jobs=-1, n_iter = 10).fit(x_train_normalized_df,y_train).best_estimator_
```

```python
ETC_pred2=Random_inputs.predict(x_test_normalized_df)
print(classification_report(y_test,ETC_pred2, digits = 4))
```
